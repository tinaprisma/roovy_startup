```python
blackbird_id = blackbird['id']
```

```python
# extract information about users that have liked restaurant
url = 'https://api.foursquare.com/v2/venues/{}/likes?&client_id={}&client_secret={}&v={}'.format(
        blackbird_id,
        client_ID,
        client_secret,
        version)

likes = requests.get(url).json()

# get number of likes
likes_count = likes['response']['likes']['count']

```

```python
# get user ids for users who liked restaurant.
likes_users_id=[]
for index in range(0,likes_count):
    user = likes['response']['likes']['items'][index]['id']
    likes_users_id.append(user)

likes_users_id

```

```python
# extract names and other info from users who have liked the restaurant
first_name =[]
last_name= []
gender= []
city = []
bio = []
for user in likes_users_id:
    url = 'https://api.foursquare.com/v2/users/{}?&client_id={}&client_secret={}&v={}'.format(
        user,
        client_ID,
        client_secret,
        version)
    user_details = requests.get(url).json()
    first_name.append(user_details['response']['user']['firstName'])
    last_name.append(user_details['response']['user']['lastName'])
    gender.append(user_details['response']['user']['gender'])
    city.append(user_details['response']['user']['homeCity'])
    bio.append(user_details['response']['user']['bio'])

data = {'First_Name': first_name, 'Last_Name': last_name, 'Gender': gender, 'City': city, 'Bio': bio}
pd.DataFrame(data=data)
```
